```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))]
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['Date'],df['Close'])
plt.xlabel('Date')
plt.ylabel('Close')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler

numerical_columns = ['Open', 'High', 'Low','Close','Adj Close','Volume']

# Separate numerical columns and datetime column
numerical_data = df[numerical_columns]
datetime_data = df['Date']

# Scale numerical data
scaler = StandardScaler()
scaled_numerical_data = scaler.fit_transform(numerical_data)

# Combine scaled numerical data with the datetime column
df_scaled = pd.concat([pd.DataFrame(scaled_numerical_data, columns=numerical_columns), datetime_data], axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
df_scaled.head(20)
```

```python
df_scaled.isnull().sum()
```

```python
df_scaled.fillna(df_scaled.mean(), inplace=True)
```

```python
df_scaled.isna().sum()
```

```python
df_scaled=df_scaled.drop('Date',axis=1)
```

```python
x=df_scaled.drop(['Close'],axis=1)
y=df_scaled['Close']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
```

```python
lin_reg.fit(x_train,y_train)
```

```python
r2_score=lin_reg.score(x_test,y_test)
print('r2 score:',r2_score)
mse=mean_squared_error(y_test,lin_reg.predict(x_test))
print('mse:',mse)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso(alpha=0.1)
```

```python
lasso.fit(x_train,y_train)
```

```python
r2_score=lasso.score(x_test,y_test)
print('r2 score:',r2_score)
mse=mean_squared_error(y_test,lasso.predict(x_test))
print('mse:',mse)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
ela_net=ElasticNet(random_state=0)
```

```python
ela_net.fit(x_train,y_train)
```

```python
r2_score=ela_net.score(x_test,y_test)
print('r2 score:',r2_score)
mse=mean_squared_error(y_test,ela_net.predict(x_test))
print('mse:',mse)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
r2_score = rf.score(x_test, y_test)
print('r2 score:', r2_score)
mse = mean_squared_error(y_test, rf.predict(x_test))
print('mse:', mse)
```

```python
from sklearn.svm import SVR
```

```python
svm=SVR(kernel='linear')
```

```python
svm.fit(x_train,y_train)
```

```python
r2_score=svm.score(x_test,y_test)
print('r2 score:',r2_score)
mse=mean_squared_error(y_test,svm.predict(x_test))
print('mse:',mse)
```

```python
models = ['Linear Regression', 'Lasso', 'Elastic Net', 'Random Forest', 'SVR']
r2_scores = [r2_score, r2_score, r2_score, r2_score, r2_score]
mse_scores = [mse, mse, mse, mse, mse]
df_results = pd.DataFrame({'Model': models, 'R2 Score': r2_scores, 'MSE': mse_scores})
df_results
```
