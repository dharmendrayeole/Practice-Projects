```python
# Label encoding on status

from sklearn.preprocessing import LabelEncoder
df['status'] = LabelEncoder().fit_transform(df['status'])
```

```python
df.dtypes
```

```python
# Label encoder on country

df['country'] = LabelEncoder().fit_transform(df['country'])
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=8)
pca.fit(scaled_df)
pca_data = pca.transform(scaled_df)

pca_data_df = pd.DataFrame(data=pca_data, columns=[f'PC{i+1}' for i in range(pca_data.shape[1])])

explained_variance_ratios = pca.explained_variance_ratio_
print("Explained Variance Ratios:")
print(explained_variance_ratios)

cumulative_explained_variance = explained_variance_ratios.cumsum()
print("\nCumulative Explained Variance:")
print(cumulative_explained_variance)

print("\nDataFrame with PCA-Transformed Data:")
print(pca_data_df)

loadings_df = pd.DataFrame(pca.components_, columns=[f'Original_Feature_{i+1}' for i in range(pca.components_.shape[1])])

print("Loadings of Each Original Feature in Each Principal Component:")
print(loadings_df)
```

```python
x=df.drop('life_expectancy',axis=1)
y=df['life_expectancy']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
lin_reg.fit(x_train,y_train)
```

```python
y_pred = lin_reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on Test Set:", mse)

r2 = r2_score(y_test, y_pred)
print("R2 Score on Test Set:", r2)
```

```python
mse_lr= mean_squared_error(y_test, y_pca_predict)
print("Mean Squared Error on Test Set:", mse_lr)

r2_lr= r2_score(y_test, y_pca_predict)
print("R2 Score on Test Set:", r2_lr)
```
